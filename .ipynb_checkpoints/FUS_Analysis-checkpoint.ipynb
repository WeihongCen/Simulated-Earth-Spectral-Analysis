## Fourier Power Spectrum Analysis

```python
import random
import matplotlib
from matplotlib.ticker import ScalarFormatter
```

```python
def reset():
    fftInput = np.array(annualArrayFUS) #Reset to default & convert list to numpy arrays
    return fftInput
```

#### Scenario 1: Unevenly spaced time series

```python
def scenario1(fftInput, chance):
    for i in range(len(fftInput)):
        if random.random() < chance:
            fftInput[i] = fftInput[i-1]
    return fftInput
```

#### Scenario 2: Few time series samples

```python
def scenario2(fftInput, dailydp):
    fftInput = np.array(annualArrayFUS)
    counter = 0
    for i in range(len(fftInput)):
        if counter < 12 / dailydp:
            fftInput[i] = fftInput[i-1]
            counter = counter + 1
        else:
            counter = 0
    return fftInput
```

#### Scenario 3: Random fluctuations

```python
def scenario3(fftInput, chance, magnitude):
    fluctuation_chance = chance
    fluctuation_magnitude = magnitude
    randomNum = random.random()
    for i in range(len(fftInput)):
        if randomNum < fluctuation_chance/2:
            fftInput[i] = fftInput[i] + random.random()*fluctuation_magnitude
        elif randomNum > 1 - fluctuation_chance/2:
            fftInput[i] = fftInput[i] - random.random()*fluctuation_magnitude
    return fftInput
```

#### Scenario 4: Sections averaged for exposure time

```python
def scenario4(fftInput, length):
    tempMean = 0
    counter = 0

    for i in range(len(fftInput)):
        counter += 1
        if counter >= length:
            counter -= length
            for j in range(int(length)):
                try:
                    tempMean = tempMean + fftInput[i + j]
                except:
                    pass
            
            tempMean = tempMean / length
            
            for x in range(int(length)):
                try:
                    fftInput[i + x] = tempMean
                except:
                    pass
            tempMean = 0
    return fftInput
```

#### Get graph parameters

```python
def findFourier(fftInput):
    fourierTransform = np.fft.fft(fftInput)
    fourierTransform = fourierTransform[0:int(len(fftInput)/2)]
    powerSpectrum = fourierTransform
    frequency = np.arange(0, len(fftInput)/2)
    return powerSpectrum, frequency
```

#### Fourier transform graph

```python
fftInput = reset()
fftInput = scenario4(fftInput, 1)
powerSpectrum, frequency = findFourier(fftInput)

start = 1
end = 2520
dpInterval = 2

fig, ax1 = plt.subplots(figsize=(14,6))
ax1.plot(1/frequency[start:end]*len(fftInput)*dpInterval, abs(powerSpectrum)[start:end])
ax1.set_yscale('log')
ax1.set_xscale('log')
ax1.grid(True)
ax1.set_xticks([3, 6, 8, 12, 24, 120])
ax1.get_xaxis().set_major_formatter(matplotlib.ticker.ScalarFormatter())

ax1.set_title('Fourier Power Spectra')
ax1.set_xlabel('Estimated Rotational Period (hours)')

psList = abs(powerSpectrum).tolist()
psList.index(max(psList[start:end]))
```

```python
errorArray = []
compiledError = []
start = 30
end = len(fftInput)
tempMean = 0
fftInput = reset()

rangeStart = 10
rangeEnd = 20
yRange = 10

for sampleRange in range(rangeStart, rangeEnd):
    fftInput = reset()
    fftInput = scenario4(fftInput, sampleRange/10)
    tempfft = fftInput
    correctEstimate = 0
    for x in range(yRange):
        randomNum = random.random()*6
        fftInput = tempfft
        powerSpectrum, frequency = findFourier(fftInput[int(randomNum*360):int(randomNum*360 + 360)])

        psList = abs(powerSpectrum).tolist()
        estimated_days = psList.index(max(psList[start:]))
        if estimated_days > 25 and estimated_days < 35:
            correctEstimate += 1
        else:
            pass
            
    compiledError = []
    errorArray.append(correctEstimate)

plt.figure(figsize=(14,6))
plt.scatter(np.arange(0, rangeEnd - rangeStart, 1), errorArray)
plt.xticks(np.arange(0, rangeEnd - rangeStart, 5))
plt.yticks(np.arange(0, yRange + 1, 1))
plt.grid(True)
```

```python
import statistics 
from statistics import mode
```

```python
def getPeak(start, interval):
    errorArray = []
    tempMean = 0
    fftInput = reset()
    xRange = 1000
    correctDays = 0

    fftInput = scenario4(fftInput, interval)
    tempfft = fftInput
    correctEstimate = 0
    for x in range(xRange):
        randomNum = random.random()*6
        fftInput = tempfft
        powerSpectrum, frequency = findFourier(fftInput[int(randomNum*360):int(randomNum*360 + 360)])

        psList = abs(powerSpectrum).tolist()
        estimated_days = psList.index(max(psList[start:]))
        errorArray.append(estimated_days)
        if estimated_days == 30:
            correctDays += 1
            
    return errorArray, correctDays
```

```python
correctArray = []
xRange = 1000 #Number of random samples

samplingInterval = 14.5 #Sampling interval(hr)

for x in range(30):
    errorArray, correctDays = getPeak(x, samplingInterval/2) #(start, interval)    
    correctChance = correctDays/1000
    correctArray.append(correctChance)


plt.figure(figsize=(14,6))
plt.plot(correctArray)
plt.grid(True)
plt.ylabel('Probability of detecting diurnal cycle')
plt.xlabel('Fourier transform starting point')
```

```python
modeArray = []

for x in range(1, 35):
    errorArray, correctDays = getPeak(x, 8)
    modeArray.append(mode(errorArray))

plt.figure(figsize=(14,6))
plt.plot(modeArray)
plt.grid(True)
```
